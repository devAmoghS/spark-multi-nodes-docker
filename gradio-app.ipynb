```python
pd.set_option('display.max_colwidth', None)

def search_df(input_key, df):
    input_key = input_key.lower()
    temp = df[df['job_details'].str.contains(input_key)].sort_values(by='category_rank')
    temp2 = temp[temp['course_title'].str.contains(input_key)].sort_values(by='category_rank')
    return temp2[['job_details','course_title', 'levenshtein_score']]
```
